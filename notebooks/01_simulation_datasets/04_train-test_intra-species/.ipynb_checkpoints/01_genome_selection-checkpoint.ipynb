````r
#' bash job using conda env
bash_job = function(cmd, conda_env, stdout=TRUE, stderr=FALSE){
    # cmd : string; commandline job (eg., 'ls -thlc')
    # conda_env : string; conda environment name
    cmd = sprintf('. ~/.bashrc; conda activate %s; %s', conda_env, cmd)
    cmd = sprintf('-c "%s"', cmd)
    system2('bash', cmd, stdout=stdout, stderr=stderr)
}
````

# Filtering

````r
# removing abnormal assemblies
metadata = metadata %>%
    filter(scaffold_count <= 100, 
           total_gap_length < 100000,
           genome_size < 10000000,
           checkm_completeness >= 90,
           checkm_contamination < 5,
           ssu_count < 20)

metadata %>% nrow
````

````r
# filtering out UBA genomes
metadata = metadata %>%
    filter(!grepl('^UBA[0-9]+$', accession)) 

metadata %>% nrow
````

# Selecting genomes by NCBI taxID

````r
# filtering out genomes lacking species classifications
metadata_s = metadata %>%
    filter(!is.na(ncbi_species_taxid),
           ncbi_species_taxid != 'none') 

metadata_s %>% nrow %>% print
metadata_s %>% head(n=3)
````

````r
# ranking based on checkM data
metadata_s = metadata_s %>%
    group_by(ncbi_species_taxid) %>%
    mutate(checkm_completeness_rank = min_rank(-checkm_completeness),
           checkm_contamination_rank = min_rank(checkm_contamination),
           summed_rank = min_rank(checkm_completeness_rank + checkm_contamination_rank),
           min_summed_rank = min(summed_rank)) %>%
    ungroup() 

metadata_s %>% nrow %>% print
metadata_s %>% head(n=3)
````

````r
# filtering (randomly selecting up to 10 of each species taxID)
metadata_s_f = metadata_s %>%
    filter(summed_rank == min_summed_rank) %>%
    group_by(ncbi_species_taxid) %>%
    sample_n(10, replace=TRUE) %>%
    ungroup() %>%
    distinct(accession, .keep_all=TRUE) %>%
    dplyr::select(-checkm_completeness_rank, -checkm_contamination_rank, 
                  -summed_rank, -min_summed_rank)

metadata_s_f %>% nrow %>% print
metadata_s_f %>% head
````

````r
# creating unique organism names
metadata_s_f = metadata_s_f %>%
    unite(ncbi_organism_name, accession, ncbi_organism_name, sep='_', remove=FALSE) 

## duplicates?
metadata_s_f %>%
    group_by(ncbi_organism_name) %>%
    summarize(n=n()) %>%
    ungroup() %>%
    filter(n>1) %>%
    nrow %>% print

# status
metadata_s_f$ncbi_organism_name %>% head(n=10) %>% print
````

## CheckM summary

### Load

````r
# Train
F = file.path(work_dir, 'GTDBr86_genome-refs_train.tsv')
metadata_train = read.delim(F, sep='\t') %>%
    mutate(data_partition = 'Train')
metadata_train %>% dim %>% print
metadata_train %>% head(n=3)
````

````r
# Test
F = file.path(work_dir, 'GTDBr86_genome-refs_test.tsv')
metadata_test = read.delim(F, sep='\t') %>%
    mutate(data_partition = 'Test')
metadata_test %>% dim %>% print
metadata_test %>% head(n=3)
````

### Summary

````r
p = metadata %>%
    dplyr::select(Taxon, data_partition, checkm_completeness, checkm_contamination) %>%
    gather(checkm_stat, checkm_value, -Taxon, -data_partition) %>%
    mutate(data_partition = factor(data_partition, levels=c('Train', 'Test')),
           checkm_stat = ifelse(checkm_stat == 'checkm_completeness', 'Completenss', 'Contamination')) %>%
    ggplot(aes(data_partition, checkm_value)) +
    geom_boxplot() +
    labs(x='Data partition', y='') +
    facet_wrap(~ checkm_stat, scales='free_y') +
    theme_bw()

options(repr.plot.width=6, repr.plot.height=3)
plot(p)
````

## Summary

````r
# loading distance matrix for train dataset
dist = fread(outF, sep='\t', header=FALSE) %>%
    mutate(V1 = basename(V1),
           V2 = basename(V2))

colnames(dist) = c('Genome.x', 'Genome.y', 'ANI', 'X', 'Y')

dist %>% dim %>% print
dist %>% head(n=3)
````

### Adding train/test labels

* which genomes belong to which dataset partition?

````r
tmp = metadata %>%
    distinct(Taxon, accession, Fasta, data_partition) %>%
    mutate(Fasta = basename(as.character(Fasta)))

dist = dist %>%
   inner_join(tmp, c('Genome.x'='Fasta')) %>%
   inner_join(tmp, c('Genome.y'='Fasta')) 

dist %>% nrow
dist %>% head
````

#### Plotting

````r
dist %>%
    group_by(data_partition.y, data_partition.x) %>%
    summarize(n = n()) %>%
    ungroup()
````

````r
p = dist %>%
    filter(Genome.x != Genome.y,
           !(data_partition.x == 'Test' & 
             data_partition.y == 'Train')) %>%
    ggplot(aes(ANI)) +
    geom_histogram(bins=50) +
    scale_y_log10() +
    labs(x='ANI', y='Count') + 
    facet_grid(data_partition.y ~ data_partition.x) +
    theme_bw()

options(repr.plot.width=5, repr.plot.height=3)
plot(p)
````
