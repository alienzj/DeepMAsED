# Goal

* Plot out DeepMAsED prediction accuracy as a function of `min(ANI)_train-test`
  * `min(ANI)_train-test` = the min ANI between the focual test genome and any genomes in the train
* Question: does accuracy decrease as phylogenetic/taxonomic novelty increases?
  * ie., can DM generalize to novel genomes?

## Functions

* Creating functions for analyzing `DeepMAsED_score ~ ANI`

**Files needed**

* DeepMAsED predictions table (created from pkl file via `unpkl_predictions.py`)
* ref-genome <--> assembly-contig index (created with `genome-contig.py`)
* ref-contig summary info (created with `merge-features.py`)
* ANI among train & test genomes (created with `fastANI`)
* ref genome metadata (from inital genome selection; GTDB)

```r
#' combining train & test ref genome metadata
join_metadata = function(train_meta_file, test_meta_file){
    train_meta = read.delim(train_meta_file, sep='\t') %>%
        mutate(data_partition = 'train',
           Fasta = basename(as.character(Fasta)),
           accession = gsub('\\.', '_', accession))
    test_meta = read.delim(test_meta_file, sep='\t') %>%
        mutate(data_partition = 'test',
           Fasta = basename(as.character(Fasta)),
           accession = gsub('\\.', '_', accession))

    meta = rbind(train_meta, test_meta) %>%
        distinct(Taxon, accession, Fasta, data_partition) %>%
        mutate(Fasta = basename(as.character(Fasta)))

    train_meta = test_meta = NULL

    return(meta)
}
```

```r
# finding max ANI of each test to any train
max_ANI_test_to_training = function(train_test_ani){
    x = train_test_ani %>%
        filter(data_partition.x == 'train',
               data_partition.y == 'test') %>%
        group_by(Genome.y, accession.y) %>%
        summarize(min_ANI = min(ANI),
                  max_ANI = max(ANI)) %>%
        ungroup() %>%
        rename('Genome' = Genome.y,
               'accession' = accession.y)

    y = train_test_ani %>%
        filter(data_partition.x == 'test',
               data_partition.y == 'train',) %>%
        group_by(Genome.x, accession.x) %>%
        summarize(min_ANI = min(ANI),
                  max_ANI = max(ANI)) %>%
        ungroup() %>%
        rename('Genome' = Genome.x,
               'accession' = accession.x)
    
    train_test_ani = rbind(x, y) %>%
        group_by(Genome, accession) %>%
        summarize(min_ANI = min(min_ANI),
                  max_ANI = max(max_ANI)) %>%
        ungroup()
    
    x = y = NULL
    return(train_test_ani)
}
```

```r
predAcc_vs_ANI = function(prediction_tbl_file, ANI_file, refGenome_asmblContig_idx_file, 
                          refContig_summary_file, train_refGenome_metadata_file,
                          test_refGenome_metadata_file, assemblers=c('megahit', 'metaspades')){
    # genome <--> contig
    genCon_idx = read.delim(refGenome_asmblContig_idx_file, sep='\t') %>%
        mutate(acc = gsub('([A-Z_[A-Z]+_[0-9]+_[0-9])_.+', '\\1', genome),
               data_type = 'test')

    cat('Genome <--> contig index, nrows:', genCon_idx %>% nrow, '\n')
    
    # ref contig summary info
    refContig_summary = fread(refContig_summary_file, sep='\t') %>%
        dplyr::select(assembler, sim_rep, contig_name, contig_length, Extensive_misassembly) %>%
        mutate(Extensive_misassembly = Extensive_misassembly != 'None')

    cat('Ref-contig summary, nrows:', refContig_summary %>% nrow, '\n')
    
    # joining genome-contig with contig-summary
    refContig_summary =  refContig_summary %>%
        inner_join(genCon_idx, c('assembler', 'sim_rep', 'contig_name'='contig')) 
    genCon_idx = NULL
    
    cat('Genome <--> contig + Ref-contig, nrows:', refContig_summary %>% nrow, '\n')
    
    # predictions
    preds = read.delim(prediction_tbl_file, sep='\t')
    cat('DeepMAsED predictions, nrows:', preds %>% nrow, '\n')
    
    # adding contig <--> genome idx to predictions
    preds = preds %>%
        inner_join(refContig_summary %>% filter(assembler %in% assemblers), 
                   c('sim_rep', 'contig'='contig_name')) 

    cat('DeepMAsED predictions + contig/genome info, nrows:', preds %>% nrow, '\n')

    # contig metadata
    meta = join_metadata(train_refGenome_metadata_file, test_refGenome_metadata_file)
    cat('Ref genome metadata, nrows:', meta %>% nrow, '\n')

    # reading ANI, joining with genome metadata
    train_test_ani = fread(ANI_file, sep='\t') %>%
        filter(V1 != V2) %>%
        dplyr::select(V1, V2, V3) %>%
        mutate(V1 = basename(V1),
               V2 = basename(V2)) %>%
        rename('Genome.x' = V1,
               'Genome.y' = V2,
               'ANI' = V3) %>%
       inner_join(meta, c('Genome.x'='Fasta')) %>%
       inner_join(meta, c('Genome.y'='Fasta')) %>%
       filter((data_partition.x == 'train' & data_partition.y == 'test') |
              (data_partition.x == 'test' & data_partition.y == 'train'))
    
    cat('Train-test ANI, nrows:', train_test_ani %>% nrow, '\n')  
    
    # finding max ANI
    train_test_ani = max_ANI_test_to_training(train_test_ani)
    cat('Train-test max ANI, nrows:', train_test_ani %>% nrow, '\n')  
 
    # ANI + predictions
    preds = preds %>%
        inner_join(train_test_ani, c('acc' = 'accession')) 

    cat('Predictions + max-ANI, nrows:', preds %>% nrow, '\n')  
    
    # ret
    return(preds)
}
```

# `score ~ ANI`

## original on n100-r30-megahit

```r
pred_file = file.path(pred_dir, 'original_n100_r30_megahit.tsv')

original_n100_r30_megahit = predAcc_vs_ANI(prediction_tbl_file = pred_file,
                                         ANI_file = train_test_ani_file,
                                         refGenome_asmblContig_idx_file = genCon_test_n100_r30_file,
                                         refContig_summary_file = conSum_test_n100_r30_file,
                                         train_refGenome_metadata_file = meta_train_n1000_file,
                                         test_refGenome_metadata_file = meta_test_n100_file,
                                         assemblers = c('megahit'))

original_n100_r30_megahit %>% dfhead
```

```r
# summarize by genome
original_n100_r30_megahit_s = original_n100_r30_megahit %>%
    group_by(genome, max_ANI, Extensive_misassembly) %>%
    summarize(mean_max_score = mean(max_score),
              stderr_max_score = sd(max_score) / sqrt(n())) %>%
    ungroup()

original_n100_r30_megahit_s %>% dfhead
```

```r
p = original_n100_r30_megahit_s %>%
    ggplot(aes(max_ANI, mean_max_score, color=Extensive_misassembly)) +
    geom_linerange(aes(ymin=mean_max_score-stderr_max_score,
                       ymax=mean_max_score+stderr_max_score),
                   alpha=0.5) +
    geom_point(alpha=0.5) +
    geom_smooth() +
    labs(x = 'Max ANI of each testing genome\nto any training genome',
         y = 'DeepMAsED score') +
    theme_bw()

dims(6, 3)
plot(p)
```

* how does genome quality (estimated by checkM) affect the score?

## Functions

```r
#' combining train & test ref genome metadata
join_metadata = function(train_meta_file, test_meta_file){
    train_meta = read.delim(train_meta_file, sep='\t') %>%
        mutate(data_partition = 'train',
           Fasta = basename(as.character(Fasta)),
           accession = gsub('\\.', '_', accession))
    test_meta = read.delim(test_meta_file, sep='\t') %>%
        mutate(data_partition = 'test',
           Fasta = basename(as.character(Fasta)),
           accession = gsub('\\.', '_', accession))

    meta = rbind(train_meta, test_meta) %>%
        distinct(Taxon, accession, Fasta, data_partition, 
                 checkm_completeness, checkm_contamination) %>%
        mutate(Fasta = basename(as.character(Fasta)))

    train_meta = test_meta = NULL

    return(meta)
}
```

```r
predAcc_vs_checkM = function(prediction_tbl_file, refGenome_asmblContig_idx_file, 
                            refContig_summary_file, train_refGenome_metadata_file,
                            test_refGenome_metadata_file, 
                            assemblers=c('megahit', 'metaspades')){
    # genome <--> contig
    genCon_idx = read.delim(refGenome_asmblContig_idx_file, sep='\t') %>%
        mutate(acc = gsub('([A-Z_[A-Z]+_[0-9]+_[0-9])_.+', '\\1', genome),
               data_type = 'test')
    cat('Genome <--> contig index, nrows:', genCon_idx %>% nrow, '\n')
    
    # ref contig summary info
    refContig_summary = fread(refContig_summary_file, sep='\t') %>%
        dplyr::select(assembler, sim_rep, contig_name, contig_length, Extensive_misassembly) %>%
        mutate(Extensive_misassembly = Extensive_misassembly != 'None')
    cat('Ref-contig summary, nrows:', refContig_summary %>% nrow, '\n')
    
    # joining genome-contig with contig-summary
    refContig_summary =  refContig_summary %>%
        inner_join(genCon_idx, c('assembler', 'sim_rep', 'contig_name'='contig')) 
    genCon_idx = NULL
    cat('Genome <--> contig + Ref-contig, nrows:', refContig_summary %>% nrow, '\n')
    
    # predictions
    preds = read.delim(prediction_tbl_file, sep='\t')
    cat('DeepMAsED predictions, nrows:', preds %>% nrow, '\n')
    
    # adding contig <--> genome idx to predictions
    preds = preds %>%
        inner_join(refContig_summary %>% filter(assembler %in% assemblers), 
                   c('sim_rep', 'contig'='contig_name')) 
    cat('DeepMAsED predictions + contig/genome info, nrows:', preds %>% nrow, '\n')

    # genome metadata
    meta = join_metadata(train_refGenome_metadata_file, test_refGenome_metadata_file)
    cat('Ref genome metadata, nrows:', meta %>% nrow, '\n')

    # predictions <--> genome metadata
    preds = preds %>%
        inner_join(meta, c('acc' = 'accession')) 
    cat('Predictions + metadata, nrows:', preds %>% nrow, '\n')  
    
    # ret
    return(preds)
}
```

```r
# finding min ANI of each test to any train
x = train_test_ani %>%
    filter(data_partition.x == 'train',
           data_partition.y == 'test') %>%
    group_by(Genome.y, accession.y) %>%
    summarize(min_ANI = min(ANI),
              max_ANI = max(ANI)) %>%
    ungroup() %>%
    rename('Genome' = Genome.y,
           'accession' = accession.y)

y = train_test_ani %>%
    filter(data_partition.x == 'test',
           data_partition.y == 'train',) %>%
    group_by(Genome.x, accession.x) %>%
    summarize(min_ANI = min(ANI),
              max_ANI = max(ANI)) %>%
    ungroup() %>%
    rename('Genome' = Genome.x,
           'accession' = accession.x)
    

train_test_ani = rbind(x, y) %>%
    group_by(Genome, accession) %>%
    summarize(min_ANI = min(min_ANI),
              max_ANI = max(max_ANI)) %>%
    ungroup()
x = y = NULL

train_test_ani %>% dfhead
```

```r
# overlap
intersect(X0_all_n100_r30_megahit_j$acc, train_test_ani$accession) %>% length
setdiff(X0_all_n100_r30_megahit_j$acc, train_test_ani$accession) %>% length
setdiff(train_test_ani$accession, X0_all_n100_r30_megahit_j$acc) %>% length
```

```r
# joining on predictions
X0_all_n100_r30_megahit_j  %>% nrow

X0_all_n100_r30_megahit_j_j = X0_all_n100_r30_megahit_j %>%
    inner_join(train_test_ani, c('acc' = 'accession')) 

X0_all_n100_r30_megahit_j_j %>% dfhead
```

```r
# summarize by genome
X0_all_n100_r30_megahit_j_j_s = X0_all_n100_r30_megahit_j_j %>%
    group_by(genome, max_ANI, Extensive_misassembly) %>%
    summarize(mean_max_score = mean(max_score),
              stderr_max_score = sd(max_score) / sqrt(n())) %>%
    ungroup()

X0_all_n100_r30_megahit_j_j_s %>% dfhead
```

```r
p = X0_all_n100_r30_megahit_j_j_s %>%
    ggplot(aes(max_ANI, mean_max_score, color=Extensive_misassembly)) +
    geom_linerange(aes(ymin=mean_max_score-stderr_max_score,
                       ymax=mean_max_score+stderr_max_score),
                   alpha=0.5) +
    geom_point(alpha=0.5) +
    geom_smooth() +
    labs(x = 'Max ANI of each testing genome\nto any training genome',
         y = 'DeepMAsED score') +
    theme_bw()

dims(6, 3)
plot(p)
```

```r
# True mis-assemblies
x = X0_all_n100_r30_megahit_j_j %>%
    filter(Extensive_misassembly == TRUE) 
x %>% nrow %>% print

cor.test(x$max_score, x$max_ANI, 
         alternative='greater',
         method = 'spearman')

## True mis-assemblies w/ ANI >80
x = X0_all_n100_r30_megahit_j_j %>%
    filter(Extensive_misassembly == TRUE,
           max_ANI > 80)
x %>% nrow %>% print

cor.test(x$max_score, x$max_ANI, 
         alternative='greater',
         method = 'spearman')

## True mis-assemblies w/ ANI <=80
x = X0_all_n100_r30_megahit_j_j %>%
    filter(Extensive_misassembly == TRUE,
           max_ANI <= 80)
x %>% nrow %>% print

cor.test(x$max_score, x$max_ANI, 
         alternative='greater',
         method = 'spearman')

# True assemblies
x = X0_all_n100_r30_megahit_j_j %>%
    filter(Extensive_misassembly == FALSE) 
x %>% nrow %>% print

cor.test(x$max_score, x$max_ANI, 
         alternative='greater',
         method = 'spearman')
```
